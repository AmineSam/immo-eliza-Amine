```python
import pandas as pd

# Load the uploaded file
df = pd.read_excel("../data/clean/immo_national.xlsx")

# Keep only actual residential property types (NL names in the dataset)
valid_types_nl = [
    "gewone woonhuizen",   # normal houses
    "appartementen",       # apartments
    "flats",               # flats
]

# Filter to only house + apartment records
df_res = df[df["CD_TYPE_NL"].str.lower().isin(valid_types_nl)]

# Extract numeric columns
num_cols = df_res.select_dtypes(include="number").columns

# Compute the minimum for each numeric column
min_values = df_res[num_cols].min()

min_values
```
